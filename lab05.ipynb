## Завдання 1

Розгляньте групу матриць за множенням  

$$G = \left\langle \left(\begin{array}{rrr}
1 & 0 & 1 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{array}\right),  \left(\begin{array}{rrr}
1 & 0 & 0 \\
0 & 1 & 1 \\
0 & 0 & 1
\end{array}\right)\right\rangle$$

Дайте відповідь на наступні питання: 

1. _Чи є група G скінченною?_
2. _Чи є група G комутативною?_
3. _Який вигляд мають елементи групи G?_
4. _Які порядки мають елементи групи G?_ 
5. _Чому ізоморфна група G?_


__Відповіді__: 
1. Група G не є скінченною.
2. Група G є комутативною.
3. $$G = \left\langle \left(\begin{array}{rrr}
1 & 0 & n \\
0 & 1 & m \\
0 & 0 & 1
\end{array}\right)\right\rangle,$$
де n це степінь матриці A, а m це степінь матриці B.
4. Елементи групи G мають нескінченний порядок. 
5. Група G ізоморфна групі цілих чисел відносно додавання. 

```python
M1 = Matrix([[1, 0, 1],
            [0, 1, 0],
            [0, 0, 1]])

M2 = Matrix([[1, 0, 0],
            [0, 1, 1],
            [0, 0, 1]])

print(M1^2, "\n")
print(M1^3, "\n")
print(M1^432, "\n")
print(M1^-32, "\n")
print(M2^2, "\n")
print(M2^3, "\n")
print(M2^432, "\n")
print(M2^-32, "\n")

if M1 * M2 == M2 * M1:
    print("Група комутативна.", "\n")
else:
    print("Група некомутативна.", "\n")

print((M1*M2)^43, "\n")
print((M2*M1)^-45, "\n")
print((M2*M1)^5, "\n")
print((M2*M1)^345, "\n")
```

\end{array}\right), 
\left(\begin{array}{rrr}

1. _Чи є група K скінченною?_
2. _Чи є група K комутативною?_
3. _Який вигляд мають елементи групи K?_
4. _Які порядки мають елементи групи K?_ 

__Відповіді__: 
1. Група K не є скінченною.
2. Група K не є комутативною.
3. $$K = \left\{ \left(\begin{array}{rrr}
(-1)^{n} & 0 & 1 \\
0 & 1 & \frac{1}{2} \cdot n \\
0 & 0 & 1
\end{array}\right) \cdot \left(\begin{array}{rrr}
1 & 0 & b \\
0 & 1 & 0 \\
0 & 0 & 1
\end{array}\right)\right\}$$
$$K = \left\{ \left(\begin{array}{rrr}
(-1)^{n} & 0 & b \\
0 & 1 & \frac{1}{2} \cdot n \\
0 & 0 & 1
\end{array}\right)\right\}$$
4. Елементи групи K мають нескінченні порядки.

__Додаткова задача:__
Опишіть підгрупи групи K з точністю до ізоморфізму.
